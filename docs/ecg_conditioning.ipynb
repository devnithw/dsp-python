# ECG signal conditioning using SciPy

```python
import numpy as np
import csv
import matplotlib.pyplot as plt
from scipy import signal
from scipy.fftpack import fft, fftshift, ifft

plt.rcParams['figure.figsize'] = (12, 3)
plt.rcParams['font.size'] = 8

%matplotlib inline
```

```python
ecg_values = []
file = 'assets/ecg_signal.csv'

with open(file, 'r') as file:
    csv_reader = csv.reader(file)
    for row in csv_reader:
        ecg_values.append(float(row[0]))
```

```python
# Convert the list to a NumPy array
ecg = np.array(ecg_values)

plt.plot(ecg, linewidth=1.0, color='b')
plt.title('ECG Waveform with powerline noise')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.xlim(0, len(ecg) - 1)
plt.show()
```

```python
plt.plot(ecg, linewidth=1.0, color='b')
plt.title('ECG Waveform with powerline noise')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.xlim(0, 500)
plt.show()
```

```python
N = len(ecg)
X = np.fft.fft(ecg , N)
spectrum = X * np.conj(X) / N

duration = 10
dt = 10 / N

freq = 1 / (dt * N) * np.arange(N)
L = np.arange(1, np.floor(N/2), dtype='int')

plt.plot(freq[L], spectrum[L], linewidth=1.0, color='r');
plt.xlim(freq[L[0]], freq[L[-1]])
plt.show()
```
